#Menampilkan data aneh

Memastikan semua data kuantitatif tidak ada anomali (bernilai mines dan atau anomali angka)

```python
#deskripsi kelas-kelas kolom data
data_cl = pd.read_csv('spotify_songs_clean.csv')
kategorikal = ['track_id','track_name','track_artist','track_album_id','track_album_name','track_album_release_date','playlist_name','playlist_id','playlist_genre','playlist_subgenre','key','mode']
q_kontinu = ['danceability','energy','loudness','speechiness','acousticness','instrumentalness','liveness','valence','tempo']
q_diskrit = ['track_popularity','duration_ms']
```

```python
print('atribut kuantitatif kontinu')

# mencari range dari data kuantitatif kontinu
for i in range(len(q_kontinu)):
    # output nama kolom
    print(f'{i+1}.',q_kontinu[i],end=": ")
    # membuat range max dan min data
    variabel_max = data_cl[q_kontinu[i]].max()
    variabel_min = data_cl[q_kontinu[i]].min()
    # output range data
    print(variabel_min,'-',variabel_max)

print()     # membuat jeda
# mencari range dari data kuantitatif diskrit
print('atribut kuantitatif diskrit')
for i in range(len(q_diskrit)):
    # output nama kolom
    print(f'{i+1}.',q_diskrit[i],end=": ")
    # membuat range max dan min data
    variabel_max = data_cl[q_diskrit[i]].max()
    variabel_min = data_cl[q_diskrit[i]].min()
    # output range data
    print(variabel_min,'-',variabel_max)
```

Mencari data pada kolom loudness yang keluar parameternya yaitu (-60) - (0)

```python
# Menentukan rentang seharusnya
expected_range = (-60, 0)

# Filter data yang berada di luar rentang
out_of_range_data = data_cl[(data_cl['loudness'] < expected_range[0]) | (data_cl['loudness'] > expected_range[1])]

# Menampilkan data yang berada di luar rentang
print("Data yang berada di luar rentang:")
out_of_range_data
```

```python
# Menghilangkan data yang berada di luar rentang
data_cl = data_cl[(data_cl['loudness'] >= expected_range[0]) & (data_cl['loudness'] <= expected_range[1])]

# Menampilkan data setelah menghilangkan data yang berada di luar rentang
print("\nData setelah menghilangkan data yang berada di luar rentang:")
data_cl
```

```python
#Memastikan data masih berupa dataset
if isinstance(data_cl, pd.DataFrame):
    # Menyimpan dataset bersih sesuai parameter ke dalam file baru
    data_cl.to_csv('spotify_songs_clean_parameter.csv', index=False)
else:
    print("Variabel data bukanlah dataframe, terdapat kesalahan pada cleansing process.")
```
