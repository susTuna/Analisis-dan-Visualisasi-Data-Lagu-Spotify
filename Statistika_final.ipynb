```python
#Menampilkan deskripsi data
df.describe()
```

```python
#Menampilkan statistik pada data kuantitatif secara lengkap
numericData1 = ["duration_ms", "tempo", "valence", "liveness", "acousticness", "speechiness", "energy","danceability"]
for i in range(len(numericData1)):
    data = numericData1[i]
    print(str(i+1) + ". " + data)
    print("Rata-rata: " + str(df[data].mean()))
    print("Standar Deviasi: " + str(df[data].std()))
    print("Persentil: ")
    print("\t10%: " + str(df[data].quantile(0.10)))
    print("\t25%: " + str(df[data].quantile(0.25)))
    print("\t50%: " + str(df[data].quantile(0.50)))
    print("\t75%: " + str(df[data].quantile(0.75)))
    print("\t90%: " + str(df[data].quantile(0.90)))
    print("Nilai ekstrim: ")
    print("\tMax: " + str(df[data].max()))
    print("\tMin: " + str(df[data].min()))
    print()
```
